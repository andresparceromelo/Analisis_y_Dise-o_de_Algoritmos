# Documentación sobre Heaps (Montículos) en Python

Un **Heap** es una estructura de datos basada en árboles que satisface la propiedad de montículo: en un *min-heap*, el padre es siempre menor o igual que sus hijos, por lo que el elemento más pequeño siempre está en la raíz.

En Python, utilizamos el módulo `heapq` que implementa un **Min-Heap** por defecto sobre una lista estándar.

## 1. Operaciones Básicas

### Crear un Heap
Puedes transformar una lista existente en un heap en tiempo lineal O(N) o empezar con una lista vacía y añadir elementos.

```python
import heapq

# 1. Convertir una lista en heap
numeros = [10, 1, 5, 20, 2, 7]
heapq.heapify(numeros)
print("Heap después de heapify:", numeros)

# 2. Insertar elementos (heappush)
heapq.heappush(numeros, 0)
print("Heap después de insertar 0:", numeros)
```

### Extraer el elemento mínimo
La operación `heappop` siempre devuelve el elemento más pequeño, manteniendo la propiedad del montículo.

```python
minimo = heapq.heappop(numeros)
print(f"Elemento extraído: {minimo}")
print("Estado del heap:", numeros)
```

## 2. Casos de Uso Comunes

### Obtener los N elementos más grandes o más pequeños
Son muy eficientes para obtener subconjuntos extremos sin ordenar toda la lista.

```python
datos = [50, 10, 4, 30, 2, 100, 80]

print("3 más grandes:", heapq.nlargest(1,datos))
print("3 más pequeños:", heapq.nsmallest(3, datos))
```

### Implementar un Max-Heap
Como Python solo ofrece Min-Heaps, para usar un Max-Heap multiplicamos los valores por -1.

```python
max_heap = []
heapq.heappush(max_heap, -10)
heapq.heappush(max_heap, -50)
heapq.heappush(max_heap, -5)

# Al extraer, el "menor negativo" será el "mayor positivo"
mayor = -heapq.heappop(max_heap)
print(f"El mayor elemento era: {mayor}")
```

## 3. Resumen de Complejidad Temporal

| Operación | Complejidad |
| :--- | :--- |
| `heapify` (Crear) | $O(N)$ |
| `heappush` (Insertar) | $O(\log N)$ |
| `heappop` (Extraer mínimo) | $O(\log N)$ |
| `nlargest` / `nsmallest` | $O(N \log K)$ |
