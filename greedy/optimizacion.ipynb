4. Minimización de Latencia (Scheduling to Minimize Lateness)Este problema no busca maximizar tareas, sino ser lo más "responsable" posible con las fechas de entrega.Explicación:Tienes $n$ tareas. Cada tarea $i$ tiene un tiempo de ejecución $t_i$ y una fecha límite (deadline) $d_i$. Si empiezas en el tiempo $s_i$, terminas en $f_i = s_i + t_i$. La latencia es $L_i = max(0, f_i - d_i)$. El objetivo es minimizar la latencia máxima.Estrategia Greedy: Ordenar las tareas únicamente por su fecha límite ($d_i$) más temprana. Ignora qué tan larga es la tarea; lo que importa es cuál vence primero.Entradas:Lista de tareas: [(duracion, deadline), ...]Salidas:El valor de la latencia máxima alcanzada.Caso de prueba:Entrada: [(3, 6), (2, 8), (1, 9), (4, 9), (3, 14), (2, 15)]Orden Greedy: Ya están ordenadas por deadline.T1: Termina en 3 (Deadline 6) -> Latencia 0T2: Termina en 5 (Deadline 8) -> Latencia 0T3: Termina en 6 (Deadline 9) -> Latencia 0T4: Termina en 10 (Deadline 9) -> Latencia 1T5: Termina en 13 (Deadline 14) -> Latencia 0T6: Termina en 15 (Deadline 15) -> Latencia 0Salida esperada: 1 (La latencia máxima es la de la T4).

```python

#primera iteracion | start 0 -> 0+3=3 -> 3-6=-3 latencia = max(0,-3) = 0
#tiempo_actual = 3

def optimizador_latencia(tareas, si = 0):
    latencias = []
    if not tareas:
        return None
    for requerido, deadline in tareas:
        if si + requerido <= deadline:
            latencia = max(0,(si+requerido-deadline))
            latencias.append(latencia)
            si += requerido

        else:
            latencia = max(0,(si+requerido-deadline))
            latencias.append(latencia)
            si += requerido

    return print(f"la mayor latencia fue de {max(latencias)}"), latencias




# prueba = optimizador_latencia(tareas = [(3, 6), (2, 8), (1, 9), (4, 9), (3, 14), (2, 15)])
# print(prueba)
# prueba = optimizador_latencia(tareas = [(5, 10), (5, 10), (5, 10)])
# print(prueba)

prueba = optimizador_latencia(tareas = [(5, 2), (4, 5), (3, 8)])
print(prueba)

```
