Backtracking construye soluciones paso a paso y retrocede cuando detecta
que un camino no lleva a una solución válida.

Patrón fundamental:
    1. Elegir   → agregar una opción al candidato
    2. Explorar → llamada recursiva
    3. Deshacer → quitar la opción (backtrack)

La PODA es lo que diferencia backtracking de fuerza bruta.
"""

Problem 1:
you have an array of numbers that doesnt repeat.

example : [1,2,3]

return all posibles combinations

return : [],[1],[1,2],[1,3],[2,3],[1,2,3],[2],[3]


```python
def combinations(array,k):
    
    final_answer, actual_answer = [],[]
    size = len(array)
    
    def backtrack(inicio):
    
        if len(actual_answer) == k:
            final_answer.append(actual_answer[:])
            return
        for i in range(inicio,size):
            actual_answer.append(array[i])
            backtrack(i+1)
            actual_answer.pop()
    
    backtrack(0)

    return final_answer

print(combinations([1,2,3,4],3))


```

```python
#intento 1 con un for
def combinaciones_parentesis(n):
    actual_answer, final_answer = [],[]
    size = 2 * n

    
    def backtrack(inicio, parentesis_abiertos, parentesis_cerrados):

        if len(actual_answer) == size:
            final_answer.append(actual_answer[:])
            return

        for i in range(size):
            if parentesis_abiertos > parentesis_cerrados and parentesis_abiertos <= size:

                actual_answer.append("(")
                parentesis_abiertos += 1
             
            backtrack(i+1)
            if parentesis_abiertos < parentesis_cerrados and parentesis_cerrados <= size:

                actual_answer.append(")")
                parentesis_cerrados += 1
            
            if len(actual_answer) > size:
                break

            actual_answer.pop()




    backtrack("(")
    return final_answer

print(combinaciones_parentesis(2))
```

```python
#intento 2 como lo hicieron en clase
def combinaciones_parentesis(n):
    actual_answer, final_answer = [],[]
    size = 2 * n

    
    def backtrack(parentesis_abiertos, parentesis_cerrados):

        if len(actual_answer) == size:
            final_answer.append(actual_answer[:])
            return

        if parentesis_abiertos < n:
            actual_answer.append("(")

            backtrack(parentesis_abiertos + 1, parentesis_cerrados)
            actual_answer.pop()

        if parentesis_cerrados < parentesis_abiertos :
            actual_answer.append(")")
            backtrack(parentesis_abiertos, parentesis_cerrados + 1)
            actual_answer.pop()
    


    backtrack(0,0)
    return final_answer

print(combinaciones_parentesis(3))
```

```python
#variacion parentesis con corchetes
#como lo hicieron en clase
def combinaciones_parentesis_corchetes(n):
    actual_answer, final_answer = [],[]
    size = 4 * n

    
    def backtrack(parentesis_abiertos, parentesis_cerrados, corchetes_abiertos, corchetes_cerrados, pendientes):

        if len(actual_answer) == size:
            final_answer.append(actual_answer[:])
            return

        if parentesis_abiertos < n:
            actual_answer.append("(")

            backtrack(parentesis_abiertos + 1, parentesis_cerrados, corchetes_abiertos, corchetes_cerrados, pendientes+"(")
            actual_answer.pop()

        if corchetes_abiertos < n:
            actual_answer.append("[")

            backtrack(parentesis_abiertos, parentesis_cerrados, corchetes_abiertos + 1, corchetes_cerrados, pendientes+"[")
            actual_answer.pop()

        if corchetes_cerrados < corchetes_abiertos:
            if len(pendientes) > 0 and pendientes[-1] == "[":
                actual_answer.append("]")

                backtrack(parentesis_abiertos, parentesis_cerrados, corchetes_abiertos, corchetes_cerrados + 1, pendientes[:-1])
                actual_answer.pop()

        if parentesis_cerrados < parentesis_abiertos:
            if len(pendientes) > 0 and pendientes[-1] == "(":
                actual_answer.append(")")

                backtrack(parentesis_abiertos, parentesis_cerrados+1, corchetes_abiertos, corchetes_cerrados, pendientes[:-1])
                actual_answer.pop()
    


    backtrack(0,0,0,0,"")
    return final_answer

print(combinaciones_parentesis_corchetes(2))
```

```python
# Ejemplo: si actual_answer = [1, 3, 0, 2], significa:

# En la fila 0, la reina está en la columna 1.
# En la fila 1, la reina está en la columna 3.
# En la fila 2, la reina está en la columna 0.
# En la fila 3, la reina está en la columna 2.




def n_reinas(n):
    
    

    final_answer, actual_answer = [],[]
    size = n

    def is_valid(col_nueva, fila_actual):
        for fila_anterior in range(len(actual_answer)):
            
            col_vieja = actual_answer[fila_anterior]

            if col_nueva == col_vieja:
                return False

            if abs(col_vieja-col_nueva) == abs(fila_anterior - fila_actual):
                return False
        return True


    def backtrack(fila_actual):
        if fila_actual == size:
            final_answer.append(actual_answer[:])
            return

        for col in range(size):
            if is_valid(col, fila_actual):
                actual_answer.append(col)
                backtrack(fila_actual+1)
                actual_answer.pop()

    backtrack(0)
    return final_answer

print(n_reinas(4))
```
